```python
import rdflib
import csv

def ttl_to_csv(ttl_file, csv_file):
    g = rdflib.Graph()
    g.parse(ttl_file, format='ttl')

    # Verzamel alle unieke predicates als kolomnamen
    predicates = set()
    for s, p, o in g:
        predicates.add(str(p))
    predicates = sorted(predicates)

    # Organiseer triples per subject
    data = {}
    for s, p, o in g:
        subj = str(s)
        pred = str(p)
        obj = str(o)
        if subj not in data:
            data[subj] = {}
        # Combineer meerdere waarden voor dezelfde predicate met een puntkomma
        if pred in data[subj]:
            data[subj][pred] += ";" + obj
        else:
            data[subj][pred] = obj

    # Schrijf CSV
    with open(csv_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        # Header: subject + predicates
        writer.writerow(['subject'] + predicates)

        for subj, preds in data.items():
            row = [subj]
            for pred in predicates:
                row.append(preds.get(pred, ''))
            writer.writerow(row)

# nlgis file comes from "https://druid.datalegend.net/nlgis/gemeentegeschiedenis/download.ttl.gz"

nlgis_ttl = "..//NLGIS//gemeentegeschiedenis.ttl"
# Voorbeeld gebruik
ttl_to_csv(nlgis_ttl, 'output.csv')
```
